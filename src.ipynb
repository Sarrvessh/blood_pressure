```python
import os
import numpy as np

def load_ppg_files(folder_path):
    """
    Load PPG waveform data from text files in a flat directory.
    
    Args:
        folder_path (str): Path to the directory containing PPG text files.
        
    Returns:
        dict: A dictionary where keys are filenames (without extension) and values are numpy arrays of PPG signals.
    """
    ppg_data = {}

    # Iterate through all text files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".txt"):  # Only process .txt files
            file_path = os.path.join(folder_path, file_name)
            try:
                # Load the PPG signal from the text file
                signal = np.loadtxt(file_path)
                # Use the filename (without extension) as the key
                ppg_data[file_name] = signal
            except Exception as e:
                print(f"Error reading file '{file_path}': {e}")

    return ppg_data

```

```python
# Convert the dictionary to a DataFrame
ppg_df = pd.DataFrame(list(ppg_data.items()), columns=["File", "Signal"])

# Preview the data
print(ppg_df.head())
```

```python
from scipy.stats import skew, kurtosis
import numpy as np

def extract_features(signal):
    """
    Extract features from a single PPG signal.
    
    Args:
        signal (array): The raw PPG signal.
    
    Returns:
        dict: A dictionary of extracted features.
    """
    return {
        "Mean": np.mean(signal),
        "Variance": np.var(signal),
        "Skewness": skew(signal),
        "Kurtosis": kurtosis(signal),
        "Min": np.min(signal),
        "Max": np.max(signal),
        "Range": np.max(signal) - np.min(signal)
    }
```

```python
# Extract features for each signal
features = ppg_df["Signal"].apply(extract_features)

# Convert the extracted features into a DataFrame
features_df = pd.DataFrame(features.tolist())

# Add the filenames for reference
features_df["File"] = ppg_df["File"]

# Preview the features
features_df
```

```python
# 1. Extract Subject ID and Segment from Filename
features_df["subject_ID"] = features_df["File"].str.split("_").str[0]  # Extract subject ID
features_df["Segment"] = features_df["File"].str.split("_").str[1].str.replace(".txt", "")  # Extract segment
features_df
```

```python
# 2. Pivot for Multiple Features
# Pivot for 'Mean'
mean_pivot = features_df.pivot(index="subject_ID", columns="Segment", values="Mean").add_prefix("Mean_")

# Pivot for 'Variance'
variance_pivot = features_df.pivot(index="subject_ID", columns="Segment", values="Variance").add_prefix("Variance_")

# Pivot for 'Skewness'
skewness_pivot = features_df.pivot(index="subject_ID", columns="Segment", values="Skewness").add_prefix("Skewness_")

# Combine all pivoted features into one DataFrame
ppg_pivot = pd.concat([mean_pivot, variance_pivot, skewness_pivot], axis=1).reset_index()

ppg_pivot
```

```python
ppg_pivot["subject_ID"] = ppg_pivot["subject_ID"].astype(str)
Final_BP_data["subject_ID"] = Final_BP_data["subject_ID"].astype(str)
```

```python
ppg_features = pd.merge(Final_BP_data, ppg_pivot, on="subject_ID", how="inner")
ppg_features
```

```python
y_diastolic.dropna(inplace=True)
y_systolic.dropna(inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train_s, y_test_s = train_test_split(X_scaled, y_systolic, test_size=0.2, random_state=42)
X_train, X_test, y_train_d, y_test_d = train_test_split(X_scaled, y_diastolic, test_size=0.2, random_state=42)
```

```python
# Combine systolic and diastolic targets
y_train_combined = np.column_stack((y_train_s, y_train_d))
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
# Multi-output regression
multi_rf_model = MultiOutputRegressor(RandomForestRegressor(random_state=42))
multi_rf_model.fit(X_train, y_train_combined)
```

```python
# Predict systolic and diastolic BP
predictions = multi_rf_model.predict(X_test)
predictions_systolic = predictions[:, 0]  # Systolic predictions
predictions_diastolic = predictions[:, 1]  # Diastolic predictions
```

```python
# Example: Evaluate systolic predictions
print("Systolic Prediction Evaluation:")
print("MAE:", mean_absolute_error(y_test_s, predictions_systolic))
print("MSE:", mean_squared_error(y_test_s, predictions_systolic))
print("RMSE:", np.sqrt(mean_squared_error(y_test_s, predictions_systolic)))


# Example: Evaluate diastolic predictions
print("\nDiastolic Prediction Evaluation:")
print("MAE:", mean_absolute_error(y_test_d, predictions_diastolic))
print("MSE:", mean_squared_error(y_test_d, predictions_diastolic))
print("RMSE:", np.sqrt(mean_squared_error(y_test_d, predictions_diastolic)))
```
